```python
from pandas import get_dummies
import matplotlib as mpl
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import warnings
import sklearn
import keras
import scipy
import numpy
import sys
import csv
import os
```

```python
sns.set(style='white', context='notebook', palette='deep')
warnings.filterwarnings('ignore')
mpl.style.use('ggplot')
sns.set_style('white')
np.random.seed(1337)
%matplotlib inline
```

```python
# import data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# data preparation training set
train_labels = train["label"].values
train_images = train.loc[:, train.columns != "label"].values
train_images = train_images / 255.0
```

```python
# data preparation test set
test_images = test.loc[:, test.columns != "label"].values
test_images = test_images / 255.0
```

```python
# construction of the model 
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(784,)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
split_point = int(0.66 * len(train_images))

x_train = train_images[:split_point]
y_train = train_labels[:split_point]

x_test = train_images[split_point:]
y_test = train_labels[split_point:]
```

```python
#please change epochs to 5
model.fit(x_train, y_train, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)

print('Test accuracy:', test_acc)
```

```python
# prediction on the set to predict
predictions = model.predict(test_images)
np.argmax(predictions[0])
```

```python
predictions_df = pd.DataFrame(predictions)
```

```python
predictions_df.head()
```

```python
predicted_labels = []
```

```python
for index, row in predictions_df.iterrows():
    predicted_labels.append(np.argmax(row))
```

```python
predicted_labels_df = pd.DataFrame(predicted_labels)
```

```python
predicted_labels_df["ImageId"] = predicted_labels_df.index
```

```python
f = lambda x: x+1
predicted_labels_df["ImageId"] = predicted_labels_df["ImageId"].apply(f)
```

```python
predicted_labels_df.columns = ["Label", "ImageId"]
```
