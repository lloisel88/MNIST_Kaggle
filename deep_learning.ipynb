```python
from pandas import get_dummies
import matplotlib as mpl
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import warnings
import sklearn
import keras
import scipy
import numpy
import sys
import csv
import os
```

```python
sns.set(style='white', context='notebook', palette='deep')
warnings.filterwarnings('ignore')
mpl.style.use('ggplot')
sns.set_style('white')
np.random.seed(1337)
%matplotlib inline
```

```python
# import data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# data preparation training set
train_labels = train["label"].values
train_images = train.loc[:, train.columns != "label"].values
train_images = train_images / 255.0
```

```python
# data preparation test set
test_images = test.loc[:, test.columns != "label"].values
test_images = test_images / 255.0
```

```python
split_point = int(0.9 * len(train_images))

x_train = train_images[:split_point]
y_train = train_labels[:split_point]

x_test = train_images[split_point:]
y_test = train_labels[split_point:]
```

```python
N_neurons_middle = [64, 128, 256, 512, 1024]
achieved_accuracy = []
for N_neurons in N_neurons_middle:
    # construction of the model 
    model = keras.Sequential([
        keras.layers.Dense(64, input_shape=(784,)),
        keras.layers.Dense(N_neurons, activation=tf.nn.relu),
        keras.layers.Dense(10, activation=tf.nn.softmax)
    ])
    model.compile(optimizer=tf.train.AdamOptimizer(), 
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    #please change epochs to 10
    model.fit(x_train, y_train, epochs=10)
    test_loss, test_acc = model.evaluate(x_test, y_test)
    achieved_accuracy.append(test_acc)
```

```python
print('For numbers of middle layers:', N_neurons_middle, "the achieved accuracies are:", achieved_accuracy)
```

```python
print("the best accuracy for middle layers is for 512 neurons. Now let us look at the first layer by selecting 512 for the middle")
```

```python
N_neurons_first = [16, 32, 64, 128, 256]
achieved_accuracy_first = []
for N_neurons in N_neurons_first:
    # construction of the model 
    model = keras.Sequential([
        keras.layers.Dense(N_neurons, input_shape=(784,)),
        keras.layers.Dense(512, activation=tf.nn.relu),
        keras.layers.Dense(10, activation=tf.nn.softmax)
    ])
    model.compile(optimizer=tf.train.AdamOptimizer(), 
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    #please change epochs to 10
    model.fit(x_train, y_train, epochs=10)
    test_loss, test_acc = model.evaluate(x_test, y_test)
    achieved_accuracy_first.append(test_acc)
```

```python
print('For numbers of first layer:', N_neurons_first, "the achieved accuracies are:", achieved_accuracy_first)
```

```python
print("the best accuracy for first layers is for 32 neurons, with 0.9714. Now let us look add an extra layer in the middle")
```

```python
# construction of the model 
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(784,)),
    keras.layers.Dense(512, activation=tf.nn.relu),
    keras.layers.Dense(512, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
#please change epochs to 10
model.fit(x_train, y_train, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Achieved accuracy:", test_acc)
```

```python
# prediction on the set to predict
predictions = model.predict(test_images)
predictions_df = pd.DataFrame(predictions)
```

```python
predicted_labels = []
for index, row in predictions_df.iterrows():
    predicted_labels.append(np.argmax(row))
```

```python
predicted_labels_df = pd.DataFrame(predicted_labels)
predicted_labels_df["ImageId"] = predicted_labels_df.index
f = lambda x: x+1
predicted_labels_df["ImageId"] = predicted_labels_df["ImageId"].apply(f)
predicted_labels_df.columns = ["Label", "ImageId"]
```
